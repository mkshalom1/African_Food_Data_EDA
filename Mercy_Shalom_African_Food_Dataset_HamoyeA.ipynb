### Which year had the least correlation with ‘Element Code’?

```python
# Using the corr() function on the food data

food_data_corr = food_data.corr(numeric_only=True)
food_data_corr
```

```python
# Getting the least correlated year with the element code

least_corr = food_data_corr.sort_values(by = 'Element Code', ascending = True)
least_corr
```

### Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?

```python
# Selecting the two columns from the dataset, I performed a groupby operation on the "Area" column, found the sum, 
# sorted it by the year and called the first seven columns. This gave the first seven lowest sum

seventh_lowest = food_data[['Y2017', 'Area']].groupby('Area').sum().sort_values(by = 'Y2017')
seventh_lowest.head(7)
```

```python
# Using the iloc[] function to get the seventh value in the dataset

seventh_lowest = food_data[['Y2017', 'Area']].groupby('Area').sum().sort_values(by = 'Y2017').iloc[6]
seventh_lowest
```

### Perform a groupby operation on ‘Element’.  What year has the highest sum of Stock Variation?

```python
# Subsetting Element column for Stock variation

stock_rows = food_data[food_data['Element'] == 'Stock Variation']
stock_rows.head()
```

```python
# Performing groupby on the already subset 'Element' column. Sum() function is called to get the sum of stock variation per year

stock_per_year = stock_rows.groupby('Element')[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].sum()
stock_per_year
```

```python
# Extracting the year with the highest stock variation for the table above using the idxmax() method

highest_stock_year = stock_per_year.idxmax(axis=1)[0]
highest_stock_year
```

### What is the total number and percentage of missing data in 2014 to 3 decimal places?

```python
# Determining the percentage of missing data

percent_miss = (no_missing / full_data_size) * 100
percent_miss
```

### What is the total Protein supply quantity in Madagascar in 2015?
